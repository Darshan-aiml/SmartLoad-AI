```python
# 1. Import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load
data = pd.read_csv('../data/cleaned_energy_data.csv', parse_dates=['Datetime'], index_col='Datetime')

# 3. Overview
data.describe()
data.info()

# 4. Visualize trends
plt.figure(figsize=(15,6))
plt.plot(data['Global_active_power'], label='Global Active Power', color='orange')
plt.title("Hourly Energy Consumption Over Time")
plt.legend()
plt.show()


```
